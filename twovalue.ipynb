```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df=pd.read_csv("/content/Wine_Quality_Data (1).csv")
df
```

```python
sample_value=df["residual_sugar"]
sample_value
```

```python
sample_sugar_1=np.random.choice(sample_value,300)

sample_sugar_2=np.random.choice(sample_value,300)

ttest,pvalue=stats.ttest_ind(sample_sugar_1,sample_sugar_2)
ttest,pvalue
```

```python
for i in range(30):
  sample_sugar_1=np.random.choice(sample_value,300)
  sample_sugar_2=np.random.choice(sample_value,300)
  ttest,pvalue=stats.ttest_ind(sample_sugar_1,sample_sugar_2)
  if pvalue<=0.07:
    print(ttest,pvalue)
    print(" Null Hypothesis is Rejected, independent")
  else:
    print(" Null Hypothesis is Accepted, dependent")
```
